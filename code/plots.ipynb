```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in excel file
df = pd.read_excel('../output/results.xlsx')
df
```

```python
# Plot scatterplot of test_acc against training_time, colored by model_type
sns.scatterplot(data=df, x='training_time', y='test_acc', hue='model_type')
plt.title('Test accuracy vs Training time by Model type')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('../output/test_acc_vs_training_time_model_type.png', bbox_inches='tight', dpi=300)
```

```python
# Plot scatterplot of test_acc against training_time, colored by model
sns.scatterplot(data=df, x='training_time', y='test_acc', hue='model')
plt.title('Test accuracy vs Training time by Model')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('../output/test_acc_vs_training_time_model.png', bbox_inches='tight', dpi=300)
```
